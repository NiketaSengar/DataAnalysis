```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv('zomato.csv',encoding='ISO-8859-1')
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
sns.heatmap(data.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
cdata=pd.read_excel('Country-Code.xlsx')
cdata.head()
```

```python
data.columns
```

```python
finaldata=pd.merge(data,cdata,on='Country Code',how='left')
```

```python
finaldata.head()
```

```python
name=finaldata.Country.value_counts().index
```

```python
count=finaldata.Country.value_counts().values
```

```python
color=['#FFCC99', '#66B2FF', '#99FF99']
plt.pie(count[:3],labels=name[:3],explode=[0.08]*3,colors=color,autopct='%1.2f%%')
```

Zomato has maximum transaction records in India(94.39%) followed by United States(4.73%) and United Kindom(0.87%).

```python
ratings=finaldata.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'rating_count'})
```

```python
ratings
```

white => not rated,
poor(red) => 1.8-2.4,
good(yellow) => 3.5-3.9,
average(orange) => 2.5-3.4,
very good(green) => 4.0-4.4,
excellent(dark green) =>4.5

```python
sns.barplot(x='Aggregate rating',y='rating_count',data=ratings)
matplotlib.rcParams['figure.figsize'] = (16, 6)
```

Not rated count is very high

```python
sns.countplot(x='Rating color',data=ratings,palette=['black','red','orange','#FFFF00','green','#097969'])
matplotlib.rcParams['figure.figsize'] = (7, 6)
```

```python
# Filter rows where Rating color is 'White' and extract the 'Country' column
country_white_ratings = finaldata[finaldata['Rating color'] == 'White'].groupby('Country').size().reset_index()
print(country_white_ratings)
```

Indians generally don't rate the service of zomato followed by brazil(5),US(3),UK(1) according to the dataset
